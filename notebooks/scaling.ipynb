## Scaling Country

# Step 1: Count the occurrences of each country

```python
country_counts = fraud_data['country'].value_counts()
```

```python
country_counts
```

# Step 2: Create a mapping from country to frequency

```python
country_freq_mapping = country_counts.to_dict()
```

```python
country_freq_mapping
```

# Step 3: Map the country column to its frequency

```python
fraud_data['country_frequency'] = fraud_data['country'].map(country_freq_mapping)
```

# Normalize the frequencies (e.g., between 0 and 1)

```python
fraud_data['country_frequency_normalized'] = fraud_data['country_frequency'] / fraud_data['country_frequency'].max()
```

## Remove unwanted columns

```python
columns_to_remove = ['user_id','signup_time', 'device_id','ip_address','country','country_frequency']
```
